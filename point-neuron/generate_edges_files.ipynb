```python
import numpy as np
import h5py
import os
import re
```

```python
def delete_file_if_exists(filename):
    if os.path.exists(filename):
        os.remove(filename)
        print("The file {} has been deleted successfully".format(filename))
    else:
        print("The file {} does not exist!".format(filename))
```

```python
# Create new SONATA edges file based on the detailed neuron SONATA edges file
delete_file_if_exists('edges_A_point.h5')
edges_A_point = h5py.File('edges_A_point.h5', 'a')

for dataset in local_edges_A['/edges/NodeA__NodeA__chemical/0']:
    print('Dataset name: ', dataset)
    local_edges_A_dataset = '/edges/NodeA__NodeA__chemical/0/{}'.format(dataset)
    nodeA_point_dataset = '/edges/NodeA__default__chemical/0/{}'.format(dataset)
    edges_A_point.create_dataset(nodeA_point_dataset, data=local_edges_A[local_edges_A_dataset], chunks=True, maxshape=(None,))
    new_data = np.array([])
    if re.search("afferent_section_id", dataset) or re.search("afferent_segment_offset", dataset) or re.search("afferent_segment_id", dataset):
        new_data = np.array([0])
    else:
        new_data = np.array([local_edges_A[local_edges_A_dataset][0]])
    if new_data.shape[0] != 0:
        edges_A_point[nodeA_point_dataset].resize((edges_A_point[nodeA_point_dataset].shape[0] + new_data.shape[0]), axis = 0)
        edges_A_point[nodeA_point_dataset][-new_data.shape[0]:] = new_data
    print("Original data: {}".format(local_edges_A[local_edges_A_dataset][:]))
    print("New data: {}".format(edges_A_point[nodeA_point_dataset][:]))

local_edges_A_dataset = '/edges/NodeA__NodeA__chemical/source_node_id'
nodeA_point_dataset = '/edges/NodeA__default__chemical/source_node_id'
edges_A_point.create_dataset(nodeA_point_dataset, data=local_edges_A[local_edges_A_dataset], chunks=True, maxshape=(None,))
edges_A_point[nodeA_point_dataset].resize((edges_A_point[nodeA_point_dataset].shape[0] + 1), axis = 0)
edges_A_point[nodeA_point_dataset][-new_data.shape[0]:] = [0]
print("Original data: {}".format(local_edges_A[local_edges_A_dataset][:]))
print("New data: {}".format(edges_A_point[nodeA_point_dataset][:]))

local_edges_A_dataset = '/edges/NodeA__NodeA__chemical/target_node_id'
nodeA_point_dataset = '/edges/NodeA__default__chemical/target_node_id'
edges_A_point.create_dataset(nodeA_point_dataset, data=local_edges_A[local_edges_A_dataset], chunks=True, maxshape=(None,))
edges_A_point[nodeA_point_dataset].resize((edges_A_point[nodeA_point_dataset].shape[0] + 1), axis = 0)
edges_A_point[nodeA_point_dataset][-new_data.shape[0]:] = [71709498]
print("Original data: {}".format(local_edges_A[local_edges_A_dataset][:]))
print("New data: {}".format(edges_A_point[nodeA_point_dataset][:]))

local_edges_A_dataset = '/edges/NodeA__NodeA__chemical/indices/source_to_target/node_id_to_ranges'
nodeA_point_dataset = '/edges/NodeA__default__chemical/indices/source_to_target/node_id_to_ranges'
edges_A_point.create_dataset(nodeA_point_dataset, data=[[0, 1], [1, 3], [3, 3]], chunks=True, maxshape=(None,None))
# edges_A_point[nodeA_point_dataset].resize((edges_A_point[nodeA_point_dataset].shape[0]), axis = 0)
# edges_A_point[nodeA_point_dataset] = [[0, 1], [1, 3], [3, 3]]
print("Original data: {}".format(local_edges_A[local_edges_A_dataset][:]))
print("New data: {}".format(edges_A_point[nodeA_point_dataset][:]))

local_edges_A_dataset = '/edges/NodeA__NodeA__chemical/indices/source_to_target/range_to_edge_id'
nodeA_point_dataset = '/edges/NodeA__default__chemical/indices/source_to_target/range_to_edge_id'
edges_A_point.create_dataset(nodeA_point_dataset, data=[[2, 4], [0, 2], [4, 5]], chunks=True, maxshape=(None,None))
# edges_A_point[nodeA_point_dataset].resize((edges_A_point[nodeA_point_dataset].shape[0]), axis = 0)
# edges_A_point[nodeA_point_dataset] = [[2, 4], [0, 2], [4, 5]]
print("Original data: {}".format(local_edges_A[local_edges_A_dataset][:]))
print("New data: {}".format(edges_A_point[nodeA_point_dataset][:]))

edges_A_point.close()
```

```python
with h5py.File('edges_point_A.h5', 'a') as f:
    # Iterate over each group
    for top_key, group in f.items():
        #Rename all datasets in the group (pad with zeros)
        for key in group.keys():
            new_key = re.sub(r'NodeA__NodeA', r'default__NodeA', key)
            group.move(key, new_key)
```
