```python
def create_edges_file(num_gids, connections_per_cell, usecase3_edges_file_name, output_file):
    tgids = []
    sgids = []
    target_to_source_node_id_to_ranges = []
    target_to_source_range_id_edge_id = []
    source_to_target_node_id_to_ranges = []
    source_to_target_range_id_edge_id = []

    total_connections = 0
    counter_for_node_id_to_ranges = 0
    for tgid in range(num_gids):
        for conn_i in range(connections_per_cell):
            sgid = (tgid + conn_i + 1) % num_gids
            sgids.append(sgid)
            tgids.append(tgid)
            total_connections += 1
        target_to_source_range_id_edge_id.append([counter_for_node_id_to_ranges, counter_for_node_id_to_ranges+connections_per_cell])
        target_to_source_node_id_to_ranges.append([tgid, tgid+1])
        counter_for_node_id_to_ranges += connections_per_cell

    sgid_instances = {}
    for i, sgid in enumerate(sgids):
        if sgid not in sgid_instances:
            sgid_instances[sgid] = []
        sgid_instances[sgid].append(i)

    counter_for_node_id_to_ranges = 0
    for sgid in range(num_gids):
        for index in sgid_instances[sgid]:
            source_to_target_range_id_edge_id.append([index, index+1])
        source_to_target_node_id_to_ranges.append([counter_for_node_id_to_ranges, counter_for_node_id_to_ranges+len(sgid_instances[sgid])])
        counter_for_node_id_to_ranges = counter_for_node_id_to_ranges + len(sgid_instances[sgid])

    target_to_source_node_id_to_ranges.append([counter_for_node_id_to_ranges, counter_for_node_id_to_ranges])
    target_to_source_node_id_to_ranges = np.array(target_to_source_node_id_to_ranges)
    target_to_source_range_id_edge_id = np.array(target_to_source_range_id_edge_id)
    source_to_target_node_id_to_ranges = np.array(source_to_target_node_id_to_ranges)
    source_to_target_range_id_edge_id = np.array(source_to_target_range_id_edge_id)
    delete_file_if_exists(output_file)
    sscx_edges_file = h5py.File(output_file, 'a')
    local_edges_A = h5py.File(usecase3_edges_file_name, 'a')

    for dataset in local_edges_A['/edges/NodeA__NodeA__chemical/0']:
        print('Dataset name: ', dataset)
        local_edges_A_dataset = '/edges/NodeA__NodeA__chemical/0/{}'.format(dataset)
        sscx_dataset = '/edges/SSCX_neurons__SSCX_neurons__chemical/0/{}'.format(dataset)
        new_data = np.array([])
        if dataset == "afferent_section_id":
            new_data = np.zeros(len(tgids))
        elif dataset == "afferent_segment_id":
            new_data = np.zeros(len(tgids))
        elif dataset == "afferent_segment_offset":
    #         new_data = np.ones(len(local_edges_A[local_edges_A_dataset]))*0.5
            new_data = np.zeros(len(tgids))
        elif dataset == "syn_type_id":
            new_data = np.ones(len(tgids))
            for i, data in enumerate(new_data):
                if i % 10 < 2:
                    new_data[i] = 0
                else:
                    new_data[i] = 101
        else:
            new_data = np.ones(len(tgids))*local_edges_A[local_edges_A_dataset][0]
        sscx_edges_file.create_dataset(sscx_dataset, data=new_data, chunks=True, maxshape=(len(tgids),))
        print("New data: {}".format(sscx_edges_file[sscx_dataset][:]))

    local_edges_A_dataset = '/edges/NodeA__NodeA__chemical/source_node_id'
    sscx_dataset = '/edges/SSCX_neurons__SSCX_neurons__chemical/source_node_id'
    sscx_edges_file.create_dataset(sscx_dataset, data=sgids, chunks=True, maxshape=(len(sgids),))
    sscx_edges_file[sscx_dataset].attrs['node_population'] = "SSCX_neurons"
    print("Dataset name: source_node_id")
    print("New data: {}".format(sscx_edges_file[sscx_dataset][:]))

    local_edges_A_dataset = '/edges/NodeA__NodeA__chemical/target_node_id'
    sscx_dataset = '/edges/SSCX_neurons__SSCX_neurons__chemical/target_node_id'
    sscx_edges_file.create_dataset(sscx_dataset, data=tgids, chunks=True, maxshape=(len(tgids),))
    sscx_edges_file[sscx_dataset].attrs['node_population'] = "SSCX_neurons"
    print("Dataset name: target_node_id")
    print("New data: {}".format(sscx_edges_file[sscx_dataset][:]))

    local_edges_A_dataset = '/edges/NodeA__NodeA__chemical/indices/source_to_target/node_id_to_ranges'
    sscx_dataset = '/edges/SSCX_neurons__SSCX_neurons__chemical/indices/source_to_target/node_id_to_ranges'
    sscx_edges_file.create_dataset(sscx_dataset, data=source_to_target_node_id_to_ranges, chunks=True, maxshape=source_to_target_node_id_to_ranges.shape)
    print("Dataset name: indices/source_to_target/node_id_to_ranges")
    print("New data: {}".format(sscx_edges_file[sscx_dataset][:]))

    local_edges_A_dataset = '/edges/NodeA__NodeA__chemical/indices/source_to_target/range_to_edge_id'
    sscx_dataset = '/edges/SSCX_neurons__SSCX_neurons__chemical/indices/source_to_target/range_to_edge_id'
    sscx_edges_file.create_dataset(sscx_dataset, data=source_to_target_range_id_edge_id, chunks=True, maxshape=source_to_target_range_id_edge_id.shape)
    print("Dataset name: indices/source_to_target/range_to_edge_id")
    print("New data: {}".format(sscx_edges_file[sscx_dataset][:]))

    local_edges_A_dataset = '/edges/NodeA__NodeA__chemical/indices/target_to_source/node_id_to_ranges'
    sscx_dataset = '/edges/SSCX_neurons__SSCX_neurons__chemical/indices/target_to_source/node_id_to_ranges'
    sscx_edges_file.create_dataset(sscx_dataset, data=target_to_source_node_id_to_ranges, chunks=True, maxshape=target_to_source_node_id_to_ranges.shape)
    print("Dataset name: indices/target_to_source/node_id_to_ranges")
    print("New data: {}".format(sscx_edges_file[sscx_dataset][:]))

    local_edges_A_dataset = '/edges/NodeA__NodeA__chemical/indices/target_to_source/range_to_edge_id'
    sscx_dataset = '/edges/SSCX_neurons__SSCX_neurons__chemical/indices/target_to_source/range_to_edge_id'
    sscx_edges_file.create_dataset(sscx_dataset, data=target_to_source_range_id_edge_id, chunks=True, maxshape=target_to_source_range_id_edge_id.shape)
    print("Dataset name: indices/target_to_source/range_to_edge_id")
    print("New data: {}".format(sscx_edges_file[sscx_dataset][:]))

    sscx_edges_file.close()
```
